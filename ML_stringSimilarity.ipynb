```python
from transformers import AutoTokenizer, AutoModel
import torch

tokenizer = AutoTokenizer.from_pretrained("dmis-lab/biobert-base-cased-v1.1")
model = AutoModel.from_pretrained("dmis-lab/biobert-base-cased-v1.1")

term1 = input("what is your term 1? ")
term2 = input("what is your term 2? ")

inputs = tokenizer([term1, term2], return_tensors="pt", padding=True)
outputs = model(**inputs)
embeddings = outputs.last_hidden_state

# Compare embeddings
similarity = torch.cosine_similarity(embeddings[0], embeddings[1]).mean()
print(similarity)
```

#### scikit 
##### Vector Space Models
a. Term Frequency-Inverse Document Frequency (TF-IDF)
Description: Converts terms into vectors based on their frequency in a collection of documents.
Usage: Calculate similarity using cosine similarity between TF-IDF vectors.

```python
!python -m pip install scikit-learn
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

term1 = input("what is your term 1? ")
term2 = input("what is your term 2? ")

vectorizer = TfidfVectorizer()
vectors = vectorizer.fit_transform([term1, term2])
similarity = cosine_similarity(vectors[0], vectors[1])
print(f"Similarity: {similarity[0][0]:.4f}")
```

#### Transformer-Based Models
a. Universal Sentence Encoder
Description: This model encodes sentences into fixed-size vectors and is effective for various semantic similarity tasks.
Usage: Encode the terms and compute similarity between their embeddings.

```python
!python -m pip install tensorflow tensorflow-hub
```

```python
import tensorflow as tf
import tensorflow_hub as hub

model = hub.load("https://tfhub.dev/google/universal-sentence-encoder/4")

def calculate_similarity(term1, term2):
    embeddings = model([term1, term2])
    return tf.keras.losses.cosine_similarity(embeddings[0], embeddings[1]).numpy()

term1 = input("what is your term 1? ")
term2 = input("what is your term 2? ")

similarity = calculate_similarity(term1,term2)
print(f"Similarity: {similarity:.4f}")
```

####  Jaccard index  (Non-ML)
The Jaccard index, or the Jaccard similarity coefficient, measures the similarity between two sets. It is defined as the ratio of the size of the intersection of the sets to the size of the union of the sets. In other words, it is the proportion of common elements between two sets.

The Jaccard index is particularly useful when the presence or absence of elements in the sets is more important than their frequency or order. For example, it can be used to compare the similarity of two documents by considering the sets of words that appear in each document.

The Jaccard index is calculated as follows:

J(A,B) = |A  B| / |A  B|

where A and B are sets, and |A| and |B| represent the cardinality or size of the sets.

```python
import re

# Function to preprocess and tokenize text
def tokenize(text):
    # Convert to lowercase
    text = text.lower()
    # Use regex to extract words
    words = re.findall(r'\b\w+\b', text)
    return set(words)

# Function to calculate the Jaccard index
def jaccard_similarity(text1, text2):
    set1 = tokenize(text1)
    set2 = tokenize(text2)
    
    # Calculate intersection and union
    intersection = set1.intersection(set2)
    union = set1.union(set2)
    
    # Compute Jaccard index
    return len(intersection) / len(union)

# Example usage

term1 = input("what is your term 1? ")
term2 = input("what is your term 2? ")

similarity = jaccard_similarity(term1, term2)
print(f"Jaccard Similarity: {similarity:.4f}")
```

```python
import torch
```

```python
import numpy as np
from gensim.models import KeyedVectors

# Load pre-trained Word2Vec model (replace this path with your model's path)
model = KeyedVectors.load_word2vec_format('path/to/word2vec/model.bin', binary=True)

# Function to convert text to vectors using the pre-trained model
def text_to_vector(text, model):
    # Tokenize the text
    words = text.split()
    word_vectors = [model[word] for word in words if word in model]
    if not word_vectors:  # Handle the case where no words are found in the model
        return np.zeros(model.vector_size)
    # Average the word vectors to get a single vector for the text
    return np.mean(word_vectors, axis=0)

# Function to calculate the cosine similarity between two texts
def cosine_similarity(text1, text2, model):
    vec1 = torch.tensor(text_to_vector(text1, model))
    vec2 = torch.tensor(text_to_vector(text2, model))
    
    # Calculate the dot product of the texts
    dot_product = torch.matmul(text1, text2.transpose(1, 0))
    
    # Calculate the norms of the texts
    norm1 = torch.norm(text1, dim=1)
    norm2 = torch.norm(text2, dim=1)
    
    # Calculate the cosine similarity
    cosine_similarity = dot_product / (norm1 * norm2)
    
    return cosine_similarity
```

```python
# Example usage

term1 = input("what is your term 1? ")
term2 = input("what is your term 2? ")

similarity = cosine_similarity(term1,term2)
print(f"Similarity: {similarity:.4f}")
```
